```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/extracted/cleaned_translated_tables.csv')
df
```

```python
df_db_id = df[['db_id','db_id_id']]
df_db_id = df_db_id.dropna()
df_db_id.columns = ['english', 'indonesia']
df_db_id
```

```python
df_tables = df[['table_names_original','table_names_original_id']]
df_tables = df_tables.dropna()
df_tables.columns = ['english', 'indonesia']
df_tables
```

```python
df_col_names_ori = df[['column_names_original','column_names_original_id']]
df_col_names_ori.columns = ['english', 'indonesia']
df_col_names_ori
```

```python
df_combined = pd.concat([df_db_id, df_tables, df_col_names_ori], ignore_index=True)
df_combined
```

```python
dict = df_combined.set_index('english')['indonesia'].to_dict()

dict
```

```python
extracted_all_schema = pd.read_csv('../../data/extracted/extracted_all_schema_words.csv')
extracted_all_schema
```

```python
def translate_word(word, translation_dict):
    return translation_dict.get(word, word)  # Jika kata ada di dict, ambil terjemahannya, jika tidak, kembalikan kata aslinya

# Mentranslate semua kata dalam kolom yang ingin ditranslate
extracted_all_schema['translated_words'] = extracted_all_schema['unique_words'].apply(lambda x: translate_word(x, dict))

output_filepath = '../../data/extracted/translated_all_schema_words.csv'
extracted_all_schema.to_csv(output_filepath, index=False)

# Menampilkan pesan konfirmasi
print(f"Hasil terjemahan telah disimpan ke file: {output_filepath}")
```
